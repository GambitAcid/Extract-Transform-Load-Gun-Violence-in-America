```python
# Dependencies and Setup
import pandas as pd
import datetime as dt

# GVA data from 
get_GVA = "../Data/GVA13_18.csv"

# Read Gun Violence Archive Data
GVA_data = pd.read_csv(get_GVA)
```

```python
# Working copy of GVA data. 
GVA = pd.DataFrame(GVA_data)
```

```python
# Create a datetime series for later comparison
GVA['comp_date'] = pd.to_datetime(GVA['date'], format= '%Y-%m-%d')
```

```python
# Retrieve only 2016 data
GVA16 = GVA[GVA['comp_date'].dt.year == 2016]
GVA16.head(5)
```

```python
GVA16.describe
```

```python
# Take a look at data by State
GVAByState16 = GVA16.groupby(['state']).count()
GVAByState16.head(20)
```

```python
# Take a look at data by State and City ot County (recorded juristiction)
GVAByStateCity16 = GVA16.groupby(['state', 'city_or_county']).count()
GVAByStateCity16.head(20) 
```

### This might be interesting

```python
# Take a look at data by State and City ot County by worst offenders
GVAByStateCity16.sort_values(by='incident_id', ascending=False).head(20) 
```

```python
# look at summary info by month
GVAByMonth16 = GVA16.groupby(GVA['comp_date'].dt.month).count()
```
